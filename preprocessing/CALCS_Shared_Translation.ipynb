```
# running Translate API
from googleapiclient.discovery import build
service = build('translate', 'v2', developerKey=APIKEY)
```

```
in_file = '/content/drive/MyDrive/Calcs/test.txt'
out_file = '/content/drive/MyDrive/Calcs/translated/test.txt'
```

```
import csv

idx = 0
with open(in_file, 'r') as f, open(out_file, 'w') as f_translated:
    for row in f:
        if idx % 10 == 0:
            print(f"{idx} completed")
        english_sentence = row.strip()
        hindi_sentence = service.translations().list(source='en', target='hi', q=[english_sentence], model='nmt').execute()['translations'][0]['translatedText']
        f_translated.write(hindi_sentence + '\n')
        idx += 1
```

```
in_file = '/content/drive/MyDrive/Calcs/dev.txt'
out_file = '/content/drive/MyDrive/Calcs/translated/dev.txt'
```

```
import csv

idx = 0
with open(in_file, 'r') as f, open(out_file, 'w') as f_translated:
    cf = csv.reader(f, delimiter="\t", quotechar='"', quoting=csv.QUOTE_NONE, escapechar=None)
    for row in cf:
        assert(len(row) == 2)
        if idx % 10 == 0:
            print(f"{idx} completed")
        english_sentence = row[0].strip()
        hindi_sentence = service.translations().list(source='en', target='hi', q=[english_sentence], model='nmt').execute()['translations'][0]['translatedText']
        f_translated.write(hindi_sentence + '\n')
        idx += 1
```

```
in_file = '/content/drive/MyDrive/Calcs/train.txt'
out_file = '/content/drive/MyDrive/Calcs/translated/train.txt'
```

```
import csv

idx = 0
with open(in_file, 'r') as f, open(out_file, 'w') as f_translated:
    cf = csv.reader(f, delimiter="\t", quotechar='"', quoting=csv.QUOTE_NONE, escapechar=None)
    for row in cf:
        assert(len(row) == 2)
        if idx % 50 == 0:
            print(f"{idx} completed")
        english_sentence = row[0].strip()
        hindi_sentence = service.translations().list(source='en', target='hi', q=[english_sentence], model='nmt').execute()['translations'][0]['translatedText']
        f_translated.write(hindi_sentence + '\n')
        f_translated.flush()
        idx += 1
```
